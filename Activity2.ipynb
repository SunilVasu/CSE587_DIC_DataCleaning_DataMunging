```r
#Input: Kagge  Sqlite  European  Soccer  Database 
#Output: R  data  frames  persisted  in  csv  files  to facilitate easy Q&A
#Process: sqllite to R data frames to c
library(dplyr)
library(RSQLite)
library(DBI)
library(plyr)
```

```r
#Loading the db downloaded from kaggle.com
con <- dbConnect(SQLite(), dbname="Data/database.sqlite")

# list all tables
dbListTables(con)
con
```

```r
#Creating df for the required tables
player <- tbl_df(dbGetQuery(con,"SELECT * FROM player"))
team <- tbl_df(dbGetQuery(con,"SELECT * FROM team"))
player_attributes <- tbl_df(dbGetQuery(con,"SELECT * FROM player_attributes"))
team_attributes <- tbl_df(dbGetQuery(con,"SELECT * FROM team_attributes"))
match <- tbl_df(dbGetQuery(con,"SELECT * FROM match"))

#write.csv(player, file = "Data/Player_Original.csv")
#write.csv(team, file = "Data/Team_Original.csv")
```

```r
#Combining the tables player, player attributes, match. The resultant table would be used
#to answer the questions related to players.
new_player <-  player %>%
  right_join(player_attributes, by = "player_api_id")


temp_match <- match
temp_player <- player
colnames(temp_match)[56]
colnames(temp_match)[56] <- "player_api_id"
temp_player <- temp_player %>% left_join(temp_match, "player_api_id" )
new <- count(temp_player, "player_name") %>% arrange(desc(freq)) 
colnames(temp_match)[56]  <- "home_player_1"

new_player <- new_player %>%
  left_join(new, by = "player_name")

leg <- temp_player %>% select (player_name,league_id)

new_player <- new_player %>%
  left_join(leg, by = "player_name")

new_player

nrow(new_player)
ncol(new_player)
#cleaning the df.
new_player <- new_player[!is.na(new_player$overall_rating) & !is.na(new_player$positioning),]

#new_player
# Saving the file as a CVS
#new_player <- new_player %>% select (player_name,overall_rating) %>% group_by(player_name)
#new_player
#str(new_player)
#summary(new_player)
nrow(new_player)
ncol(new_player)

#storing the df as csv
write.csv(new_player, file = "Data/Player.csv")
```

```r
# 1) Which are the top 10 players by overall rating?
#new_player - retrieving the information from the created table to ans the questions.
top10 <- 
  new_player %>% select (player_name,overall_rating, positioning) %>% 
  arrange(desc(overall_rating)) %>% 
  #limit 10  %>%
  distinct(player_name,overall_rating)  %>%
  #group_by(player_name)%>%
  as.data.frame()

#top10 
#top10 <- top10 %>%  distinct(player_name) 
top10 <-  top10  %>% head(n = 10) %>%  distinct(player_name,overall_rating)
top10
```
